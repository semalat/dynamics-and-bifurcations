# Сечение Пуанкаре для разлчиных $I_0$


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from math import pi
```

```python
def ode(y, t):
        T, U, V = y
        dydt = [w, eps * (V - b * U - a), V - (V ** 3)/3 - U + I_0 * np.sin(T)]
        return dydt
```

```python
def calcODE(args, y0, dy0, ddy0, ts, nt):
    y = [y0, dy0, ddy0]
    t = np.linspace(0, ts, nt)
    sol = odeint(ode, y, t)
    return sol
```

```python
def parametr_changing(args, deltaX, deltaDX, deltaDDX, startX, stopX, startDX, stopDX, startDDX, stopDDX, ts, nt):
    y0 = 0
    dy0 = 0
    ddy0 = 0
    sol = calcODE(args, y0, dy0, ddy0, ts, nt)
    #print(len(sol[900:, 1]))
    return list(sol[900:, 1])
```

```python
#Внимание! Функция долго считает (~3 минут)



I_0 = 0.01
I_0_all = []
sol_all = []

for i in np.arange(0, 1, 0.01):
    I_0 += 0.01
    #print('I0 = ', I_0)
    w = 1;
    a = -0.7;
    b = 0.8;
    eps = 0.08
    args=(b, eps, a, w, I_0)
    sol_curr = parametr_changing(args,1, 1, 1, -6, 6, -6, 6, -6, 6, ts = 10000 * pi , nt = 1001)
    I_0_all.extend([I_0 for _ in sol_curr])
    sol_all.extend(sol_curr)
```

```python
# 100 results
plt.scatter(I_0_all, sol_all, s= 1, c = 'b')
plt.xlabel('I0')
plt.ylabel('U')
plt.grid(1)
plt.show()
```
