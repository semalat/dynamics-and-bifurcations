# Случай трех переменных. Зависимость от времени.



$\begin{equation}
			\begin{cases}
				\pmb{V_t= V- \frac{V^3}{3}- U + I_0 sin(\theta)}\\
				\pmb{\tau  U_t= V - \text{$\beta U - \alpha$}}\\
				\pmb{\theta_t = \omega t}
			\end{cases}
		\end{equation}$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def ode(y, t):
        T, U, V = y
        dydt = [w, eps * (V - b * U - a), V - (V ** 3)/3 - U + I0 * np.sin(T)]
        return dydt
```

```python
def calcODE(args, y0, dy0, ddy0, ts, nt):
    y = [y0, dy0, ddy0]
    t = np.linspace(0, ts, nt)
    sol = odeint(ode, y, t)
    return sol
```

```python
def drawPhasePortrait_t(args, deltaX, deltaDX, deltaDDX, startX, stopX, startDX, stopDX, startDDX, stopDDX, ts, nt):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    for y0 in range(startX, stopX, deltaX):
        for dy0 in range(startDX, stopDX, deltaDX):
            for ddy0 in range(startDDX, stopDDX, deltaDDX):
                sol = calcODE(args, y0, dy0, ddy0, ts, nt)
                plt.plot(sol[9000:, 2], sol[9000:, 1], sol[9000:, 0], 'b')
    plt.xlabel('V')
    plt.ylabel('U')
    plt.grid(1)
    plt.show()
```

```python
def projection_xy(args, deltaX, deltaDX, deltaDDX, startX, stopX, startDX, stopDX, startDDX, stopDDX, ts, nt):
    for y0 in range(startX, stopX, deltaX):
        for dy0 in range(startDX, stopDX, deltaDX):
            ddy0 = 0
            sol = calcODE(args, y0, dy0, ddy0, ts, nt)
            plt.plot(sol[9000:, 2], sol[9000:, 1], 'b')
    plt.xlabel('V')
    plt.ylabel('U')
    plt.grid(1)
    plt.show()
```

```python
def projection_xz(args, deltaX, deltaDX, deltaDDX, startX, stopX, startDX, stopDX, startDDX, stopDDX, ts, nt):
    for y0 in range(startX, stopX, deltaX):
        for dy0 in range(startDX, stopDX, deltaDX):
            ddy0 = 0
            sol = calcODE(args, y0, dy0, ddy0, ts, nt)
            plt.plot(sol[9000:, 1], sol[9000:, 0], 'b')
    plt.xlabel('T')
    plt.ylabel('U')
    plt.grid(1)
    plt.show()
```

```python
def projection_yz(args, deltaX, deltaDX, deltaDDX, startX, stopX, startDX, stopDX, startDDX, stopDDX, ts, nt):
    for y0 in range(startX, stopX, deltaX):
        for dy0 in range(startDX, stopDX, deltaDX):
            ddy0 = 0
            sol = calcODE(args, y0, dy0, ddy0, ts, nt)
            plt.plot(sol[9000:, 0], sol[9000:, 2], 'b')
    plt.xlabel('T')
    plt.ylabel('V')
    plt.grid(1)
    plt.show()
```

```python
I0 = 1;
w = 1;
a = -0.7;
b = 0.8;
eps = 0.08;
args=(b, eps, a)
drawPhasePortrait_t(args, 1, 1, 1, -6, 6, -6, 6, -6, 6, ts = 1000, nt = 10000)
```
